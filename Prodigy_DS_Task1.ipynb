```python
# Plotting the relationship between Region and Income Groups

data_frame = pd.DataFrame(my_info)

#number of countries in each income group for each region
income_by_region = pd.crosstab(data_frame.IncomeGroup, data_frame.Region)

# Colors
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

fig, ax = plt.subplots(figsize=(8, 6))

# Creating bar chart
income_by_region.plot(kind='bar', color=colors, ax=ax)

plt.xlabel('Income Group')
plt.ylabel('Number of Countries in the Regions')
plt.title('Income Group Distribution by Region')

# Show legend
plt.legend(title='Region')

plt.xticks(rotation=45)
plt.show()
```

So, we have successfully analyzed the relationship between the variables and have been able to visualize using Bar Charts
